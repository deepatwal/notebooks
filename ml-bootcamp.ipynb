# cost functions

## example 1 - a simple cost function

## $$ f(x) = x^2 + x + 1 $$

```python
def f(x):
    return x**2 + x + 1
```

```python
x_1 = np.linspace(start=-3, stop=3, num=100)
```

```python
x_1
```

```python
plt.xlim([-3, 3])
plt.ylim([0, 8])
plt.xlabel("x", fontdict={"fontsize": 16})
plt.ylabel("f(x)", fontdict={"fontsize": 16})
plt.plot(x_1, f(x_1))
plt.show()
```

## slope & derivatives
- slope of a function with one variable is a rate of change and is calculated by the derivative of a function

```python
def df(x):
    return 2*x + 1
```

## plot example 1 function and its derivative

```python
plt.figure(figsize=[15, 5])

plt.subplot(1, 2, 1)
plt.xlim(-3, 3)
plt.ylim(0, 8)
plt.xlabel("x", fontdict={"fontsize": 16})
plt.ylabel("f(x)", fontdict={"fontsize": 16})
plt.title("cost function")
plt.plot(x_1, f(x_1))

plt.subplot(1, 2, 2)
plt.xlim(-2, 3)
plt.ylim(-3, 6)
plt.xlabel("x", fontdict={"fontsize": 16})
plt.ylabel("df(x)", fontdict={"fontsize": 16})
plt.title("slope of the cost function")
plt.plot(x_1, df(x_1))
plt.grid()

plt.show()
```
