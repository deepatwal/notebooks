# Import

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

# Example 1: Cost Function

## $$ f(x) = x^2 + x + 1 $$

```python
def f(x):
    return x**2 + x + 1
```

```python
# data for example 1

x_1 = np.linspace(-3, 3, 500)
```

```python
plt.plot(x_1, f(x_1))
plt.xlim([-3, 3])
plt.ylim([0, 9])
plt.xlabel('x', fontdict={'size': 15})
plt.ylabel('f(x)', fontdict={'size': 15})
plt.show()
```

# Example 1: Slope & Derivative

## $$ df(x) = 2x + 1 $$

```python
def df(x):
    return 2*x + 1
```

```python
# plotting f(x) and its derivative df(x) side by side

plt.figure(figsize=[15,5])

plt.subplot(1, 2, 1)
plt.title('cost function')
plt.xlim([-3, 3])
plt.ylim([0, 9])
plt.xlabel('x', fontdict={'size': 15})
plt.ylabel('f(x)', fontdict={'size': 15})
plt.plot(x_1, f(x_1))
plt.grid()

plt.subplot(1, 2, 2)
plt.title('slope of a cost function')
plt.xlim([-2, 3])
plt.ylim([-3, 6])
plt.xlabel('x', fontdict={'size': 15})
plt.ylabel('df(x)', fontdict={'size': 15})
plt.plot(x_1, df(x_1), color='skyblue', linewidth=4)
plt.grid()

plt.show()
```
