# Example 1: Cost Function

## $$ f(x) = x^2 + x + 1 $$

```python
def f(x):
    return x**2 + x + 1
```

```python
# data for example 1

x_1 = np.linspace(-3, 3, 500)
```

```python
plt.plot(x_1, f(x_1))
plt.xlim([-3, 3])
plt.ylim([0, 9])
plt.xlabel('x', fontdict={'size': 15})
plt.ylabel('f(x)', fontdict={'size': 15})
plt.show()
```

# Example 1: Slope & Derivative

## $$ df(x) = 2x + 1 $$

```python
def df(x):
    return 2*x + 1
```

```python
# plotting f(x) and its derivative df(x) side by side

plt.figure(figsize=[15,5])

plt.subplot(1, 2, 1)
plt.title('cost function')
plt.xlim([-3, 3])
plt.ylim([0, 9])
plt.xlabel('x', fontdict={'size': 15})
plt.ylabel('f(x)', fontdict={'size': 15})
plt.plot(x_1, f(x_1))
plt.grid()

plt.subplot(1, 2, 2)
plt.title('slope of a cost function')
plt.xlim([-2, 3])
plt.ylim([-3, 6])
plt.xlabel('x', fontdict={'size': 15})
plt.ylabel('df(x)', fontdict={'size': 15})
plt.plot(x_1, df(x_1), color='skyblue', linewidth=4)
plt.grid()

plt.show()
```

```python
def gradient(initial=0, learning_rate=0.001, precision=0.00001, iteration=1):
    
    new_x = initial
    previous_x = 0
    x_list = []
    gradients = []

    for i in range(iteration):
        previous_x = new_x
        gradient_ = df(previous_x)
        new_x = previous_x - gradient_ * learning_rate

        x_list.append(previous_x)
        gradients.append(gradient_)

        step_size = abs(new_x - previous_x)
        if step_size <= precision:
            print(f"step size is less than the precision value")
            break

    x_list.append(new_x)
    gradients.append(df(new_x))

    print(f"local minimum occur at: {new_x}")
    print(f"cost or f(x) value for local minimum: {f(new_x)}")
    print(f"slope or df(x) value for local minimum: {df(new_x)}")
    
    return x_list, gradients
```

```python
x_list, gradients = gradient(initial=2, learning_rate=0.1, iteration=100)
```

```python
# plotting the gradient descent

plt.figure(figsize=[15, 5])

plt.subplot(1, 2, 1)
plt.title('cost function')
plt.xlim([-3, 3])
plt.ylim([0, 9])
plt.xlabel('x', fontdict={'size': 15})
plt.ylabel('f(x)', fontdict={'size': 15})

plt.plot(x_1, f(x_1))
plt.scatter(x_list, f(np.array(x_list)), color="red")
plt.grid()

plt.subplot(1, 2, 2)
plt.title('slope of a cost function')
plt.xlim([-2, 3])
plt.ylim([-3, 6])
plt.xlabel('x', fontdict={'size': 15})
plt.ylabel('df(x)', fontdict={'size': 15})

plt.plot(x_1, df(x_1), color='skyblue',)
plt.scatter(x_list, df(np.array(x_list)), color="red")
plt.grid()
```
